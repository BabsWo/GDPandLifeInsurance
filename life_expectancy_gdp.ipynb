# GDP and Life Expectancy a CodeCademy project

The first thing to do is identify goals. The goals as stated by the project overview is to analyze the relationship between GDP and life expectancy for six countries. If possible, I would also like to assess inequities by race or gender identification in this data. After looking at the data, I determined that there was no gender or race information stored in the data set, but that the data has a time component. Therefore, it will be necessary to assess how this data changes with time. The data seems to span from 2000 to 2015. 


```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("all_data.csv")
print(df.head(25))
```

```python
df.rename(columns = {"Life expectancy at birth (years)": "Life_expectancy"}, inplace= True)
```

```python
sns.set_palette("colorblind")
sns.set_context("paper")
sns.lineplot(data = df, x = "Year", y="GDP", hue="Country", markers=True)
plt.title("GDP by country per year")
```

```python
chile = df[df.Country == "Chile"]
china = df[df.Country == "China"]
germany = df[df.Country == "Germany"]
mexico = df[df.Country == "Mexico"]
us = df[df.Country == "United States of America"]
zimbabwe = df[df.Country == "Zimbabwe"]

```

```python
fig, axes = plt.subplots(2, 3)
fig.tight_layout(pad=3)
plt.subplot(2,3,1)
sns.lineplot(data = chile, x="GDP", y="Life_expectancy")
plt.title("Chile")
plt.subplot(2,3,2)
sns.lineplot(data = china, x="GDP", y="Life_expectancy")
plt.title("China")
plt.subplot(2,3,3)
sns.lineplot(data = germany, x="GDP", y="Life_expectancy")
plt.title("Germany")
plt.subplot(2,3,4)
sns.lineplot(data = mexico, x="GDP", y="Life_expectancy")
plt.title("Mexico")
plt.subplot(2,3,5)
sns.lineplot(data = us, x="GDP", y="Life_expectancy")
plt.title("United States")
plt.subplot(2,3,6)
sns.lineplot(data = zimbabwe, x="GDP", y="Life_expectancy")
plt.title("Zimbabwe")

```

Clearly there is a relationship between GDP and Life Expectancy across these countries. Plotting the data like this, where each country is isolated, helps us understand that because it doesn't obscure the relationship. Plotting them on top of each other would lead to issues because the overall GDP for each country is very different.
